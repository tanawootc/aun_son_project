```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() 
```

```python
x = np.array(d1T.index).reshape(-1, 1)
y = np.array(d1T[0]).reshape(-1, 1)
```

```python
model.fit( x,y)
```

```python
predit = model.predict([[800]]) 
```

```python
predit
```

```python
intercept = model.intercept_
print(intercept)
coefficient = model.coef_
print(coefficient[0]) 
```
