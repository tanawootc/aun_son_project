```python
from sklearn.linear_model import Perceptron
import pandas as pd
df = pd.read_excel("data_normolrized2.xlsx",header=1)
```

```python
#display(df)

x = df[['P410','P435','P460','P485']]
y = df['No.']
model = Perceptron()
model.fit(x, y)
predict = model.predict([[10.827236	,7.463653	,100.808916	,5.190092]])
print('predict:', predict[0])
print('weights:', model.coef_)
print('bias:', model.intercept_)
```
